```r
# Instalar pacotes necessários
if(!require("tidyverse")) install.packages("tidyverse")
if(!require("viridis")) install.packages("viridis")

# Carregar bibliotecas
library(tidyverse)
library(viridis)

# Dataset como string
data_string <- "ID,Group,Age,Gender,Programming_Experience,Anxiety_State,Self_Efficacy,Condition_LLM,Condition_Blend,task_1,task_2,task_3,task_4,task_5,task_6,task_7,task_8,task_9,task_10,Eye_Fixation_Duration,Eye_Fixation_Count,Eye_Saccade_Amplitude,Eye_Pupil_Diameter
1,1,20,1,2,35,85,1,1,7.2,7.5,7.3,7.8,7.1,7.6,7.4,7.2,7.9,7.5,280,140,5.8,3.9
2,1,22,2,1,32,88,1,1,7.5,7.8,7.6,8.1,7.4,7.9,7.7,7.5,8.2,7.8,250,125,5.4,3.6
3,1,19,1,3,38,83,1,1,7.0,7.3,7.1,7.6,6.9,7.4,7.2,7.0,7.7,7.3,300,150,6.2,4.2
4,1,21,2,2,30,90,1,1,7.8,8.1,7.9,8.4,7.7,8.2,8.0,7.8,8.5,8.1,230,115,5.0,3.3
5,1,18,1,1,37,86,1,1,7.3,7.6,7.4,7.9,7.2,7.7,7.5,7.3,8.0,7.6,270,135,5.6,3.8
6,2,23,1,2,40,82,1,0,6.8,7.1,6.9,7.4,6.7,7.2,7.0,6.8,7.5,7.1,320,160,6.6,4.5
7,2,20,2,1,37,85,1,0,7.1,7.4,7.2,7.7,7.0,7.5,7.3,7.1,7.8,7.4,290,145,6.0,4.1
8,2,18,1,3,39,80,1,0,6.7,7.0,6.8,7.3,6.6,7.1,6.9,6.7,7.4,7.0,340,170,7.0,4.8
9,2,22,2,2,35,87,1,0,7.4,7.7,7.5,8.0,7.3,7.8,7.6,7.4,8.1,7.7,260,130,5.5,3.7
10,2,19,1,1,41,83,1,0,7.0,7.3,7.1,7.6,6.9,7.4,7.2,7.0,7.7,7.3,310,155,6.4,4.4
11,3,21,2,1,38,80,0,1,6.8,7.1,6.9,7.4,6.7,7.2,7.0,6.8,7.5,7.1,330,165,6.8,4.7
12,3,18,1,2,36,83,0,1,7.0,7.3,7.1,7.6,6.9,7.4,7.2,7.0,7.7,7.3,300,150,6.2,4.2
13,3,22,2,3,40,78,0,1,6.6,6.9,6.7,7.2,6.5,7.0,6.8,6.6,7.3,6.9,350,175,7.2,4.9
14,3,19,1,2,34,85,0,1,7.2,7.5,7.3,7.8,7.1,7.6,7.4,7.2,7.9,7.5,270,135,5.6,3.8
15,3,23,2,1,39,81,0,1,6.9,7.2,7.0,7.5,6.8,7.3,7.1,6.9,7.6,7.2,320,160,6.6,4.5
16,4,20,2,2,42,77,0,0,6.5,6.8,6.6,7.1,6.4,6.9,6.7,6.5,7.2,6.8,370,185,7.6,5.2
17,4,22,1,1,40,80,0,0,6.7,7.0,6.8,7.3,6.6,7.1,6.9,6.7,7.4,7.0,340,170,7.0,4.8
18,4,19,2,3,41,75,0,0,6.4,6.7,6.5,7.0,6.3,6.8,6.6,6.4,7.1,6.7,390,195,8.0,5.5
19,4,21,1,2,38,82,0,0,6.9,7.2,7.0,7.5,6.8,7.3,7.1,6.9,7.6,7.2,310,155,6.4,4.4
20,4,18,2,1,43,78,0,0,6.6,6.9,6.7,7.2,6.5,7.0,6.8,6.6,7.3,6.9,360,180,7.4,5.1
21,1,21,1,1,33,87,1,1,7.4,7.7,7.5,8.0,7.3,7.8,7.6,7.4,8.1,7.7,260,130,5.5,3.7
22,2,19,2,2,37,84,1,0,7.1,7.4,7.2,7.7,7.0,7.5,7.3,7.1,7.8,7.4,280,140,5.8,3.9
23,3,23,1,3,35,86,0,1,7.5,7.8,7.6,8.1,7.4,7.9,7.7,7.5,8.2,7.8,240,120,5.2,3.5
24,4,20,2,0,40,79,0,0,6.6,6.9,6.7,7.2,6.5,7.0,6.8,6.6,7.3,6.9,330,165,6.8,4.7
25,1,18,2,2,32,89,1,1,7.7,8.0,7.8,8.3,7.6,8.1,7.9,7.7,8.4,8.0,220,110,4.9,3.2
26,2,22,1,1,38,83,1,0,7.0,7.3,7.1,7.6,6.9,7.4,7.2,7.0,7.7,7.3,290,145,6.0,4.1
27,3,19,2,0,36,85,0,1,7.3,7.6,7.4,7.9,7.2,7.7,7.5,7.3,8.0,7.6,250,125,5.4,3.6
28,4,21,1,1,39,80,0,0,6.8,7.1,6.9,7.4,6.7,7.2,7.0,6.8,7.5,7.1,320,160,6.6,4.5
29,1,20,2,3,31,90,1,1,7.8,8.1,7.9,8.4,7.7,8.2,8.0,7.8,8.5,8.1,210,105,4.7,3.1
30,2,22,1,2,36,85,1,0,7.2,7.5,7.3,7.8,7.1,7.6,7.4,7.2,7.9,7.5,270,135,5.6,3.8
31,3,23,2,1,34,88,0,1,7.6,7.9,7.7,8.2,7.5,8.0,7.8,7.6,8.3,7.9,230,115,5.1,3.4
32,4,19,1,0,41,77,0,0,6.5,6.8,6.6,7.1,6.4,6.9,6.7,6.5,7.2,6.8,350,175,7.2,4.9
33,1,21,1,2,37,87,1,1,7.4,7.7,7.5,8.0,7.3,7.8,7.6,7.4,8.1,7.7,250,125,5.4,3.6
34,2,18,2,1,35,84,1,0,7.1,7.4,7.2,7.7,7.0,7.5,7.3,7.1,7.8,7.4,270,135,5.7,3.8
35,3,22,1,3,33,86,0,1,7.4,7.7,7.5,8.0,7.3,7.8,7.6,7.4,8.1,7.7,240,120,5.2,3.5
36,4,20,2,2,38,81,0,0,6.8,7.1,6.9,7.4,6.7,7.2,7.0,6.8,7.5,7.1,310,155,6.5,4.4
37,1,19,2,0,36,83,1,1,7.0,7.3,7.1,7.6,6.9,7.4,7.2,7.0,7.7,7.3,280,140,5.8,3.9
38,2,23,1,1,34,86,1,0,7.3,7.6,7.4,7.9,7.2,7.7,7.5,7.3,8.0,7.6,260,130,5.5,3.7
39,3,21,2,2,32,88,0,1,7.6,7.9,7.7,8.2,7.5,8.0,7.8,7.6,8.3,7.9,230,115,5.1,3.4
40,4,18,1,0,40,79,0,0,6.6,6.9,6.7,7.2,6.5,7.0,6.8,6.6,7.3,6.9,340,170,7.0,4.8"

# Ler o dataset
df <- read.csv(text = data_string)

# Normalizar variáveis neurofisiológicas para evitar escalas desiguais
df_normalized <- df %>%
  mutate_at(vars(Eye_Fixation_Duration, Eye_Fixation_Count, Eye_Saccade_Amplitude, Eye_Pupil_Diameter), scale)

# Adicionar rótulos de condições experimentais
df_normalized$LLM_Label <- factor(ifelse(df$Condition_LLM == 1, "Com LLM", "Sem LLM"))
df_normalized$Blend_Label <- factor(ifelse(df$Condition_Blend == 1, "Com Blend", "Sem Blend"))

# Definir um tema alternativo para os gráficos
theme_set(theme_minimal() + theme(plot.background = element_rect(fill = "#f0f0f0")))

# Função para gerar gráficos de densidade para variáveis neurofisiológicas
gerar_grafico_densidade <- function(var1, var2) {
  ggplot(df_normalized, aes(x = .data[[var1]], fill = Blend_Label)) +
    geom_density(alpha = 0.7) +
    facet_wrap(~ LLM_Label) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Comparação de", var1, "e", var2),
         x = var1,
         y = "Densidade") +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Função para gerar boxplots comparando neurofisiologia e escalas
gerar_grafico_caixa <- function(var1, var2) {
  ggplot(df_normalized, aes(x = Blend_Label, y = .data[[var1]], fill = Blend_Label)) +
    geom_boxplot(alpha = 0.7) +
    facet_wrap(~ LLM_Label) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Boxplot de", var1, "e", var2),
         x = "Condição Blend",
         y = var1) +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Geração de gráficos: comparações de duas escalas ou neurofisiologia
variables <- c("Anxiety_State", "Self_Efficacy", "Eye_Fixation_Duration", "Eye_Fixation_Count", "Eye_Saccade_Amplitude", "Eye_Pupil_Diameter")

graficos_densidade <- list()
graficos_caixa <- list()

# Gerar gráficos de densidade e boxplots para pares de variáveis
for (i in 1:(length(variables)-1)) {
  for (j in (i+1):length(variables)) {
    graficos_densidade[[paste(variables[i], variables[j], sep = "vs")]] <- gerar_grafico_densidade(variables[i], variables[j])
    graficos_caixa[[paste(variables[i], variables[j], sep = "vs")]] <- gerar_grafico_caixa(variables[i], variables[j])
  }
}

# Exibir os gráficos
print(graficos_densidade)
print(graficos_caixa)
```
